```python
Payroll_df = pd.read_csv("Citywide_Payroll_Data__Fiscal_Year_.csv", low_memory=False)
Payroll_df.head()
```

```python
Salary_Small = Payroll_df[["Agency Name","Base Salary"]]
Salary_Small.head()
```

```python
New_Base_Salary = Salary_Small[Salary_Small.columns[1:]].replace('[\$,]', '', regex=True).astype(float)
Salary_Small['Base Salary'] = New_Base_Salary
Salary_Small.head()
```

```python
Group_Payroll = Salary_Small.groupby(by='Agency Name').agg({'Base Salary': 'sum'}).reset_index()
Group_Payroll1 = Group_Payroll.head(15)
#.set_index("Agency Name", inplace = True)
Group_Payroll1
```

```python
#@Roaya TOP NUMBER OF BASE SALARY PER AGENCY
Base_Top = Group_Payroll1.sort_values(by=["Base Salary"], ascending=False)
Base_Top.head(20)
```

```python
Payroll_Small = Payroll_df[["Agency Name","OT Hours", "Total OT Paid"]]
Payroll_Small.head()
```

```python
Total_Pay_OT = Payroll_Small[Payroll_Small.columns[2]].replace('[\$,]', '', regex=True).astype(float)
Payroll_Small['Total OT Paid'] = Total_Pay_OT
Payroll_Small.head()
```

```python
Group_OT = Payroll_Small.groupby(by='Agency Name').agg({'OT Hours': 'sum', 'Total OT Paid': 'sum'}).reset_index()
Group_OT.head(10)
```

```python
#@Roaya TOP OT HOURS and PAID PER AGENCY
OT_Hours_Top = Group_OT.sort_values(by=["OT Hours"], ascending=False)
OT_Hours_Top.head(20)
```

```python
OT_Total_Top = Group_OT.sort_values(by=["Total OT Paid"], ascending=False)
OT_Total_Top.head()
```
